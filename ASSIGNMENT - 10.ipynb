```python
#convert the large data into wide data format using pivot
book_df = book.pivot_table(index='UserID',columns='Title',values='Rating').reset_index(drop=True)
book_df
```

```python
#replace the index value by unquie userid
book_df.index = book['UserID'].unique()
book_df
```

```python
#fill the nan value with 0
book_df.fillna(0,inplace=True)
book_df
```

```python
#Calculate the cosine similarity score
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation

user = pairwise_distances(book_df,metric='cosine')
user
```

```python
#store the result in dataframe format
user_df = pd.DataFrame(user)
user_df
```

```python
#Getting the recommendation of the data
def get_recommendation(customer_id):
        tem=list(user_df.sort_values([customer_id],ascending=False).head(3).index)
        book_list=list(book[book['UserID'].isin(tem)]['Title'])
        recommended_books = set(book_list)-set(book[book['UserID']==customer_id]['Title'])
        return recommended_books
```
